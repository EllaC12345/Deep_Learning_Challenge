## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.columns = application_df.columns.str.strip()
application_df.drop(['EIN', 'NAME'], axis=1, inplace = True)
```

```python
# Determine the number of unique values in each column.
unique_value_count = application_df.nunique()
unique_value_count
```

```python
# Look at APPLICATION_TYPE value counts for binning
APPLICATION_TYPE_count = application_df['APPLICATION_TYPE'].value_counts()
APPLICATION_TYPE_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`


# Set your chosen cutoff value
cutoff_value = 500

value_counts = application_df['APPLICATION_TYPE'].value_counts()

# Create a list of application types to be replaced with 'Other'
application_types_to_replace = value_counts[value_counts < cutoff_value].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_count = application_df['CLASSIFICATION'].value_counts()
classification_count
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_count_over_one = classification_count[classification_count > 1]
classification_count_over_one
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# Set your chosen cutoff value
cutoff_value = 500

classification_count = application_df['CLASSIFICATION'].value_counts()

# Create a list of application types to be replaced with 'Other'
classifications_to_replace = classification_count[classification_count < cutoff_value].index.tolist()


# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
## splitting target from dataset
y =application_df['IS_SUCCESSFUL']
X = application_df.drop(columns = 'IS_SUCCESSFUL')
X.head(5)
y.head()
```

```python
X.head()
```

```python
cons = X["SPECIAL_CONSIDERATIONS"].unique()
cons
```

```python
status = X["STATUS"].unique()
status
```

```python
# Encode booloean variables
def consideration(SPECIAL_CONSIDERATIONS):
  if SPECIAL_CONSIDERATIONS == "Y":
    return 1
  else:
      return 0
# Call the encode_fee function on the fee column
X["SPECIAL_CONSIDERATIONS"] = X["SPECIAL_CONSIDERATIONS"].apply(consideration)

```

```python
# Encode the categorical variables using get_dummies
X_transformed= pd.get_dummies(X[["APPLICATION_TYPE", "AFFILIATION", "CLASSIFICATION", "USE_CASE", "ORGANIZATION", "INCOME_AMT"]])

## Display the transformed data
X_transformed.head()

```

```python
# Concatenate the X_transformed and the X DataFrames
X = pd.concat([X, X_transformed], axis=1)
# Drop the original  columns
X = X.drop(columns=["APPLICATION_TYPE", "AFFILIATION", "CLASSIFICATION", "USE_CASE", "ORGANIZATION", "INCOME_AMT"])

X.head()
```

```python
# Split our preprocessed data into our features and target arrays

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train,y_test = train_test_split(X,y, random_state = 1)
X_train.shape
X_test.shape
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
## determining # of input and neuron for modeling
input = X.columns.nunique()
input


```

## Compile, Train and Evaluate the Model

```python
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp):
    nn_model = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh','sigmoid'])

    # Allow kerastuner to decide number of neurons in first layer
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=10,
        step=2), activation=activation, input_dim=input))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 6)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=10,
            step=2),
            activation=activation))

    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])

    return nn_model
```

```python
# Import the kerastuner library
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=3)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer based on the results of hyperparameter tuner
hidden_nodes_layer1 = 7
hidden_nodes_layer2 = 3
hidden_nodes_layer3 = 3
hidden_nodes_layer4 = 3
hidden_nodes_layer5 = 3
hidden_nodes_layer6 = 1

nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(
    tf.keras.layers.Dense(units= hidden_nodes_layer1, input_dim=input, activation = 'relu')
)


# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units= hidden_nodes_layer2, input_dim=input, activation = 'relu')
       )
# Third hidden layer
nn_model.add(tf.keras.layers.Dense(units= hidden_nodes_layer3, input_dim=input, activation = 'relu')
       )

# Fourth hidden layer
nn_model.add(tf.keras.layers.Dense(units= hidden_nodes_layer4, input_dim=input, activation = 'relu')
       )

# Fith hidden layer
nn_model.add(tf.keras.layers.Dense(units= hidden_nodes_layer5, input_dim=input, activation = 'relu')
       )

# Sixth hidden layer
nn_model.add(tf.keras.layers.Dense(units= hidden_nodes_layer6, input_dim=input, activation = 'relu')
       )

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model= nn_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
